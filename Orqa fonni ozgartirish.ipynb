# Background replacement

```python
%matplotlib inline
import numpy as np
import cv2
import matplotlib.image as mimg
import matplotlib.pyplot as plt
rasm=cv2.imread('green girl.png')
rasm=cv2.cvtColor(rasm,cv2.COLOR_BGR2RGB)
plt.imshow(rasm)
plt.show()
rasm.shape
```

```python
copg=np.copy(rasm)
gir=np.array([0,142,0])
gir2=np.array([[150,255,150]])
maska=cv2.inRange(copg,gir,gir2)
plt.imshow(maska,cmap='gray')
plt.show()
```

```python
copg[maska!=0]=0
plt.imshow(copg)
plt.show()
```

```python
car=cv2.imread('car.jpg')
car=cv2.cvtColor(car,cv2.COLOR_BGR2RGB)
car=car[:720,:1156]

car[maska==0]=[0,0,0]
plt.imshow(car)
plt.show()
car.shape
```

```python
%matplotlib inline
tayyor=car+copg
plt.imshow(tayyor)
plt.show()
```
